```python
x = df['Happiness.Score']
y = df['Economy..GDP.per.Capita.']
plt.scatter(x,y , s=5)
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"grey")


plt.xlabel('Happiness Score')
plt.ylabel('Gross Domestic Product')
plt.title("Country Happiness Score v's GDP")
```
