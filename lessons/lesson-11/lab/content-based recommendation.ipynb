```python
import pandas
```

```python
songs = pandas.read_csv('../../data/songs.csv')
listeners = pandas.read_csv('../../data/listeners.csv')
listeners
```

```python
songs
```

```python
len(songs.columns)
```

```python
juanita = listeners[listeners.juanita.notnull()]
juanita_df = juanita.merge(songs, left_on='name', right_on='name')
juanita_df
```

```python
column_names = juanita_df.columns[3:]
column_names
```

```python
import sklearn.linear_model
X = juanita_df[column_names]
y = juanita_df.juanita
elastic = sklearn.linear_model.ElasticNetCV()
elastic.fit(X,y)
```

```python
elastic.coef_
```

```python
sorted([(score, attribute) for (score, attribute) in zip(elastic.coef_, column_names)
  if abs(score) > 0.1])
```

```python
elastic.intercept_
```

```python
# what would she like?
what_we_think_she_will_rate = elastic.predict(songs[column_names])
what_we_think_she_will_rate
```

```python
juanita_prediction = pandas.DataFrame({
        'song_name' : songs.name,
        'juanita_score': what_we_think_she_will_rate
        }
                 )
juanita_prediction
```

```python
juanita_prediction.loc[juanita_prediction.juanita_score.idxmax()]
```
